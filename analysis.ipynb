```python
#Task 2:
# Check if there are any missing values
df.isnull().sum()
```

```python
# Total missing values
print("Total Missing Values :", df.isnull().sum().sum())
```

```python
# Drop columns that are not useful
df.drop(["EmployeeNumber",
         "EmployeeCount",
         "Over18",
         "StandardHours"],
        axis=1,
        inplace=True)
print("New Shape :", df.shape)
```

```python
# Convert Attrition column
df["Attrition"] = df["Attrition"].map({
    "Yes":1,
    "No":0
})
df["Attrition"].head()
```

```python
print(df["Attrition"].value_counts())
```

```python
# Numerical columns
num_cols = df.select_dtypes(include=["int64","float64"]).columns

# Categorical columns
cat_cols = df.select_dtypes(include=["object"]).columns
print("Numerical Columns :", len(num_cols))
print("Categorical Columns :", len(cat_cols))
```

```python
# Convert categorical columns into dummy variables
df = pd.get_dummies(df,
                    columns=cat_cols,
                    drop_first=True)
print("Shape after Encoding :", df.shape)
```

```python
df.head()
```

```python
# Features
X = df.drop("Attrition", axis=1)

# Target
y = df["Attrition"]
print("X Shape :", X.shape)
print("y Shape :", y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Convert scaled data into DataFrame
X_scaled = pd.DataFrame(X_scaled,
                        columns=X.columns)
X_scaled.head()
```

```python
print("Final Feature Shape :", X_scaled.shape)
print("Target Shape :", y.shape)
```

```python
#Task 4:
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
# Split dataset into training and testing data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
print("Training Data :", X_train.shape)
print("Testing Data :", X_test.shape)
```

```python
# Create Logistic Regression model
lr_model = LogisticRegression(
    class_weight="balanced",
    random_state=42
)
# Train the model
lr_model.fit(X_train, y_train)

# Prediction
lr_pred = lr_model.predict(X_test)
```

```python
lr_accuracy = accuracy_score(y_test, lr_pred)
lr_precision = precision_score(y_test, lr_pred)
lr_recall = recall_score(y_test, lr_pred)
lr_f1 = f1_score(y_test, lr_pred)

print("Accuracy :", lr_accuracy)
print("Precision :", lr_precision)
print("Recall :", lr_recall)
print("F1 Score :", lr_f1)
```

```python
# Create Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,
    class_weight="balanced",
    random_state=42
)
# Train model
rf_model.fit(X_train, y_train)
# Prediction
rf_pred = rf_model.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_precision = precision_score(y_test, rf_pred)
rf_recall = recall_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred)

print("Accuracy :", rf_accuracy)
print("Precision :", rf_precision)
print("Recall :", rf_recall)
print("F1 Score :", rf_f1)
```

```python
# Create Gradient Boosting model
gb_model = GradientBoostingClassifier(
    random_state=42
)
# Train model
gb_model.fit(X_train, y_train)

# Prediction
gb_pred = gb_model.predict(X_test)
```

```python
gb_accuracy = accuracy_score(y_test, gb_pred)
gb_precision = precision_score(y_test, gb_pred)
gb_recall = recall_score(y_test, gb_pred)
gb_f1 = f1_score(y_test, gb_pred)

print("Accuracy :", gb_accuracy)
print("Precision :", gb_precision)
print("Recall :", gb_recall)
print("F1 Score :", gb_f1)
```

```python
# Create comparison table
result = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    
    "Accuracy": [
        lr_accuracy,
        rf_accuracy,
        gb_accuracy
    ],
    
    "Precision": [
        lr_precision,
        rf_precision,
        gb_precision
    ],
    
    "Recall": [
        lr_recall,
        rf_recall,
        gb_recall
    ],
    
    "F1 Score": [
        lr_f1,
        rf_f1,
        gb_f1
    ]
})

result
```

```python
best_model = result.sort_values(
    by="F1 Score",
    ascending=False
)
best_model
```

#Observation :-
1. Three machine learning models were trained for predicting employee attrition.
2. Logistic Regression provides a good baseline and performs well on balanced datasets.
3. Random Forest captures complex patterns and generally gives higher accuracy.
4. Gradient Boosting also performs well and often gives strong prediction results.
5. The model with the highest F1 Score is selected as the best model because F1 Score balances both Precision and Recall.

# Task 7: HR Insights & Business Recommendations
1. Which 3 factors most strongly predict that an employee will leave?
-> From the analysis and the machine learning model, the three most important factors related to employee attrition are **OverTime**, **Monthly Income**, and **Years at Company**. Employees who work overtime frequently are more likely to leave the organization. Lower monthly income is also associated with a higher attrition rate. In addition, employees who have spent fewer years in the company are more likely to resign compared to employees who have been with the organization for a longer period.

2. Which department or job role should HR prioritize for retention efforts?
->The analysis shows that the **Sales department** has a higher number of employees leaving the company. Among different job roles, **Sales Representatives** and **Laboratory Technicians** have the highest attrition rates. HR should give special attention to these employees by providing better career growth opportunities, regular feedback sessions, and employee engagement programs.

3. Does salary alone explain attrition, or are there other stronger factors?
->No, salary alone does not explain why employees leave the company. Although employees with lower salaries tend to leave more often, other factors such as overtime, years at the company, work-life balance, and job role also have a significant impact. This shows that employee attrition is influenced by multiple factors rather than salary alone.

4. Two HR recommendations based on the findings
**Recommendation 1:**
HR should regularly monitor employees who work overtime or have recently joined the company. Conducting one-on-one meetings, reducing excessive workload, and providing mentoring during the initial years can help improve employee retention.

**Recommendation 2:**
The company should focus on employees working in high-attrition departments such as Sales. Providing competitive salaries, clear promotion opportunities, and regular skill development programs can improve job satisfaction and reduce employee turnover.

5. What limitation does this model have?
->This model is built using historical employee data, so it cannot capture personal reasons for leaving, such as family situations, better job opportunities, or relationships with managers. The model should be used as a decision-support tool rather than the only factor in HR decisions. Final decisions should always consider employee feedback and managerial judgment along with the model's predictions.
